```python
# Week 3 Assignment: Exploratory Data Analysis (EDA)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load cleaned dataset
df = pd.read_csv("TeslaStock_Dataset_Cleaned.csv")

# Step 2: General overview
print("Dataset Overview:")
display(df.head(10))

# Step 3: Plot 1 - Closing Price over Time
plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['Close'], color='blue')
plt.title("Tesla Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price ($)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
print("Insight: Tesla's closing price shows general trends, peaks, and dips over time.")

# Step 4: Plot 2 - Volume Traded over Time
plt.figure(figsize=(12,5))
plt.bar(df['Date'], df['Volume'], color='orange')
plt.title("Tesla Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.show()
print("Insight: Volume spikes may indicate high trading activity during certain events.")

# Step 5: Plot 3 - High vs Low Prices
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['High'], label='High', color='green')
plt.plot(df['Date'], df['Low'], label='Low', color='red')
plt.title("Tesla High vs Low Prices")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
print("Insight: The gap between high and low prices indicates daily volatility.")

# Step 6: Plot 4 - Closing Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Close'], bins=30, kde=True, color='purple')
plt.title("Distribution of Tesla Closing Prices")
plt.xlabel("Close Price ($)")
plt.ylabel("Frequency")
plt.show()
print("Insight: Most closing prices fall within a certain range, showing common trading levels.")

# Step 7: Plot 5 - Correlation Heatmap (numeric columns only)
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include='number')  # select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Tesla Stock Features")
plt.show()
print("Insight: Close, High, Low, and Open prices are highly correlated, while volume shows weaker correlation.")
```
