```python
# Week 4 Assignment: Correlation Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load cleaned dataset
df = pd.read_csv("TeslaStock_Dataset_Cleaned.csv")

# Step 2: Check dataset overview
print("Dataset Overview:")
display(df.head(10))

# Step 3: Select numeric columns
numeric_df = df.select_dtypes(include='number')

# Step 4: Correlation matrix
correlation_matrix = numeric_df.corr()
print("\nCorrelation Matrix:")
display(correlation_matrix)

# Step 5: Plot heatmap of correlations
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Tesla Stock Features")
plt.show()

# Step 6: Find top 3 features most correlated with target (Closing Price)
target = 'Close'
correlations_with_target = correlation_matrix[target].drop(target)  # remove self-correlation
top3_features = correlations_with_target.abs().sort_values(ascending=False).head(3)
print(f"\nTop 3 features most related to {target}:")
print(top3_features)
```
