```python
# Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned Titanic dataset
df = pd.read_csv("titanic_cleaned.csv")

# Set style
sns.set(style="whitegrid")

# ----------------- Plot 1: Survival Count -----------------
plt.figure(figsize=(6,4))
sns.countplot(x="survived", data=df)
plt.title("Plot 1: Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

print("Insight 1: More passengers did not survive (0) than those who survived (1).")

# ----------------- Plot 2: Survival by Gender -----------------
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="survived", data=df)
plt.title("Plot 2: Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

print("Insight 2: Females had a higher survival rate than males, showing the 'women and children first' evacuation rule.")

# ----------------- Plot 3: Age Distribution -----------------
plt.figure(figsize=(6,4))
sns.histplot(df["age"], bins=30, kde=True, color="blue")
plt.title("Plot 3: Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

print("Insight 3: Most passengers were between 20–40 years old. Very few were elderly.")

# ----------------- Plot 4: Class vs Fare -----------------
plt.figure(figsize=(6,4))
sns.boxplot(x="pclass", y="fare", data=df)
plt.title("Plot 4: Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()

print("Insight 4: 1st class passengers paid much higher fares than 2nd and 3rd class passengers.")

# ----------------- Plot 5: Embarkation Port vs Survival -----------------
plt.figure(figsize=(6,4))
sns.countplot(x="embarked", hue="survived", data=df)
plt.title("Plot 5: Survival by Embarkation Port")
plt.xlabel("Embarkation Port")
plt.ylabel("Count")
plt.show()

print("Insight 5: Passengers who boarded from port 'C' (Cherbourg) had a higher survival rate compared to other ports.")
```
