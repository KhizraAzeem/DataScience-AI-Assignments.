```python
# Load Titanic dataset from seaborn
df = sns.load_dataset("titanic")

# Show first 10 rows (original)
print("=== First 10 rows (original) ===")
df.head(10)
```

```python
print("=== Basic info (before cleaning) ===")
print("Columns:", list(df.columns))
print("Shape:", df.shape)
print("\nMissing values per column:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
print("\nNumber of duplicate rows:", df.duplicated().sum())


```

```python
# Make a copy so original df remains for comparison
df_cleaned = df.copy()

# Columns to drop (Seaborn Titanic has these extra columns compared to Kaggle)
cols_to_drop = ["deck", "embark_town", "alive", "adult_male", "who", "class"]

# Drop only if they exist (safe drop)
cols_present = [c for c in cols_to_drop if c in df_cleaned.columns]
if cols_present:
    df_cleaned = df_cleaned.drop(columns=cols_present)

# Fill missing numeric 'age' with median (if exists)
if "age" in df_cleaned.columns:
    df_cleaned["age"].fillna(df_cleaned["age"].median(), inplace=True)

# Fill missing 'embarked' with mode (if exists)
if "embarked" in df_cleaned.columns:
    if df_cleaned["embarked"].isnull().sum() > 0:
        df_cleaned["embarked"].fillna(df_cleaned["embarked"].mode()[0], inplace=True)

# Fill missing 'embark_town' if present (safety)
if "embark_town" in df_cleaned.columns:
    if df_cleaned["embark_town"].isnull().sum() > 0:
        df_cleaned["embark_town"].fillna(df_cleaned["embark_town"].mode()[0], inplace=True)

# Remove duplicate rows (if any) and reset index
duplicates_before = df_cleaned.duplicated().sum()
df_cleaned.drop_duplicates(inplace=True)
df_cleaned.reset_index(drop=True, inplace=True)

print(f"Cleaning done. Dropped columns: {cols_present}. Duplicates removed: {duplicates_before - df_cleaned.duplicated().sum()}")
```

```python
print("=== After cleaning summary ===")
print("Shape:", df_cleaned.shape)
print("\nMissing values per column after cleaning:\n", df_cleaned.isnull().sum())
print("\nNumber of duplicate rows after cleaning:", df_cleaned.duplicated().sum())

print("\n=== First 10 rows (cleaned) ===")
df_cleaned.head(10)
```

```python
def shape_str(t):
    return f"{t[0]} rows x {t[1]} cols"

report = pd.DataFrame({
    "Before Cleaning": [shape_str(df.shape), int(df.isnull().sum().sum()), int(df.duplicated().sum())],
    "After Cleaning":  [shape_str(df_cleaned.shape), int(df_cleaned.isnull().sum().sum()), int(df_cleaned.duplicated().sum())]
}, index=["Shape", "Missing Values (total)", "Duplicates"])

print("=== Before vs After Cleaning Report ===")
report
```

```python
csv_filename = "titanic_cleaned.csv"
df_cleaned.to_csv(csv_filename, index=False)
print(f"✅ Cleaned dataset saved as: {csv_filename}")

```
